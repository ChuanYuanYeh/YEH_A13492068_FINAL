```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players = pd.read_csv('Players.csv')
seasons = pd.read_csv('Seasons_Stats.csv')
```

```python
players.drop('Unnamed: 0', axis=1, inplace=True)
players.dropna(how='all', inplace=True)
players.head()
```

```python
seasons.drop('Unnamed: 0', axis=1, inplace=True)
seasons['Mnutes/Game'] = seasons.MP/seasons.G
seasons['Points/Game'] = seasons.PTS/seasons.G
seasons['Assists/Game'] = seasons.AST/seasons.G
seasons['Rebounds/Game'] = seasons.TRB/seasons.G
seasons['Steals/Game'] = seasons.STL/seasons.G
seasons['Blocks/Game'] = seasons.BLK/seasons.G
seasons.head()
```

```python
seasons_allen = seasons.loc[seasons.Player == 'Ray Allen']
seasons_allen.to_csv('seasons_allen.csv')
```

```python
sgs = seasons.loc[seasons.Pos == 'SG'].copy()
sgs.head()
```

```python
sgs.head()
```

```python
allen_stats = seasons_allen[['Points/Game','Assists/Game','Rebounds/Game','Steals/Game','Blocks/Game']].mean()
sgs_stats = sgs.groupby('Player')[['Points/Game','Assists/Game','Rebounds/Game','Steals/Game','Blocks/Game']].apply(np.mean).dropna(how='any').mean()
radar_data = pd.DataFrame({'All Shooting Guards Career Avg.': sgs_stats,
             'Ray Allen Career Avg.': allen_stats})
radar_data.to_csv('radarData.csv')
```
